```python
from twython import Twython
import pandas as pd
```

```python
CONSUMER_KEY = '...' #You need to enter your consumer key here.
CONSUMER_SECRET = '...' #You need to enter your consumer secret here.

twitter = Twython(CONSUMER_KEY, CONSUMER_SECRET)
```

```python
results=twitter.search(q="bottle",count=100)
all_items=results["statuses"]
df=pd.DataFrame(all_items)
df.head()
#We pull our data about bottle from Twitter.
```

```python
users=pd.DataFrame(d["user"] for d in results["statuses"])
users.head()
```

```python
tdf=df.merge(users,left_index=True,right_index=True)
tdf.info()
```

```python
tdf.location.value_counts().plot(kind="bar",figsize=(10,6))
```

```python
from textblob import TextBlob
```

```python
df2=tdf[["text"]]
df2.head()
```

```python
def detect_sentiment(text):
    return TextBlob(text).sentiment.polarity
#We are doing our sentiment analysis.
```

```python
df2["sentiment"]=df2["text"].apply(detect_sentiment)
df2.head()
```

```python
df2.sentiment.value_counts()
```

```python
df2["text"]=df2["text"].str.lower() #We convert our texts to lowercase.
df2["text"]=df2["text"].str.replace("[^\w\s]","") #We remove punctuation marks from our texts.
df2["text"]=df2["text"].str.replace("\d+","") #We are removing numbers from our texts.
df2["text"]=df2["text"].str.replace("\n","").replace("\r","") #We remove spaces in our texts.
```

```python
from wordcloud import STOPWORDS
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
def wc(data,bgcolor):
    plt.figure(figsize=(10,10))
    mask1=np.array(Image.open("bottle.png"))
    wc=WordCloud(background_color=bgcolor,stopwords=STOPWORDS,mask=mask1)
    wc.generate(" ".join(data))
    plt.imshow(wc)
    plt.axis("off")
#We paint the most frequently used words in tweets on a bottle.
```

```python
wc(df2.text,"black")
```

```python
def sentiment2(sent):
    if (sent< -0.02):
        return 3
    elif sent>0.02:
        return 1
    else:
        return 0
#We divide the texts into three groups positive, negative and n√∂tr.
```

```python
df2["sent"]=df2["sentiment"].apply(sentiment2)
df2.head()
```

```python
import seaborn as sns
sns.countplot(df2.sent)
```

```python
x=df2.text
y=df2.sent
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
vect=CountVectorizer(stop_words="english")
x=vect.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.15)
```

```python
mb=MultinomialNB()
tahmin=mb.fit(x_train,y_train).predict(x_test)
accuracy_score(tahmin,y_test)
```

```python
d=DecisionTreeClassifier()
tahmin=d.fit(x_train,y_train).predict(x_test)
accuracy_score(tahmin,y_test)
```
